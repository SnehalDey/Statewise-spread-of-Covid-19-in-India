```python
#converting date-time to ordinal
MH['date']=MH['date'].map(dt.datetime.toordinal)
MH.head()
```

```python
# importing random forest for prediciton
from sklearn.ensemble import RandomForestRegressor

# selecting  the dependet and independt variable for prediction
y = MH['confirmed']

x=MH.drop(columns=["state",'confirmed'],axis=1)
```

```python
# reshaping the y
y = y.values.reshape(-1,1)
```

```python
# splitting the train and test data
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.3)
```

```python
# using the random forest model and tuning the hyperparameters
reg = RandomForestRegressor(n_estimators=3, random_state=10, min_samples_split=2, n_jobs=-1)
```

```python
# training the model
reg.fit(train_x,train_y)
```

```python
# predicitng the results
pred = reg.predict(train_x)
prediction = reg.predict(test_x)
```

```python
# calculating the RMSE
from sklearn import metrics
rmsecm = float(format(np.sqrt(metrics.mean_squared_error(train_y, pred)),'.3f'))
print(rmsecm)
```
